```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
a= pd.read_csv('2015.csv')
a.head()
```

```python
b= pd.read_csv('2016.csv')
b.head()
```

```python
c= pd.read_csv('2017.csv')
c.head()
```

```python
d= pd.read_csv('2018.csv')
d.head()
```

```python
e= pd.read_csv('2019.csv')
e.head()
```

```python
#concatenate the datasets into one
datasets= pd.concat([a,b,c,d,e])
datasets.head()
```

```python
datasets.describe()
```

```python
datasets.columns
```

```python
datasets.head()
# The Happiest Country in the world is Switzerland with a (HappinessScore = 7.587, \
#EconomyGDP =1.39, FreedomChoice score = 0.66, LifeExpectancy=0.9, GovernmentCorruption = 0.41
```

```python
datasets.tail()
# 4 Countries with low Happiness score are Rwanda, Tanzania, Afghanistan,Central African Republic and South Sudan\.
#  The factors that contribute to this are:

# Low LifeExpectancy
# Low Freedom of choice or exercise of human right
# Low EconomyGDP
```

```python
correlation =datasets.corr()
f, ax = plt.subplots(figsize = (20, 20))
sns.heatmap(correlation, annot = True)

# 5 Positive Contributing Factors To Happiness In a Country
# HappinessScore is strongly correlated with Economy (GDP per Capita)
# HappinessScore is strongly correlated with Family Bond
# HappinessScore is strongly correlated with Health   (High LifeExpectancy)
# HappinessScore is strongly correlated with  (High SocialSupport)
# HappinessScore is correlated with   (High Freedom of Choice)

# 4 Negative Contributing Factors To Happiness In a Country
# HappinessScore is least correlated with GovernmentCorruption meaning
# HappinessScore is poorly correlated with  Generosity
# HappinessScore is least correlated with DystopiaResidual
# HappinessScore is least correlated with  StandardError
```

```python
sns.lmplot(data = datasets, x = "HappinessScore", y = "HappinessScore",fit_reg=True,\
                  hue = "Year",\
                 size = 6, aspect= 1.5, scatter_kws = {'s':20}, )
 
#From here we, can see that there is an increase in World's Happiness by year
#as its shwing currently 2019 is the highest
```
